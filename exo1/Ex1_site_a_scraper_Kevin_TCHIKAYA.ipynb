import des éléments 

```python
from IPython.display import Image, display
from urllib.parse import urljoin
import requests
from bs4 import BeautifulSoup
```

url du site à scrper

```python
url = "https://pythonprogramming.net/parsememcparseface/"
response = requests.get(url, timeout=10)
response.raise_for_status()
```

Soup

```python
soup = BeautifulSoup(response.text, "html.parser")
#print(soup.get_text())
```

**Récupérer le petit tableau**

```python
petit_tableau = [
    element.get_text(strip=True)
    for element in soup.find_all("li")
]

for liste in soup.find_all("ul"):
    valeurs = [li.get_text(strip=True) for li in liste.find_all("li", recursive=False)]

    if "Python" in valeurs and "Pascal" in valeurs:
        petit_tableau = valeurs
        break

print(petit_tableau)
```

**Récupérer le grand tableau**

```python
grand_tableau = []

for ligne in soup.find("table").find_all("tr"):
    cellules = ligne.find_all(["th", "td"])
    grand_tableau.append([
        cellule.get_text(" ", strip=True)
        for cellule in cellules
    ])

print(grand_tableau)
```

**Récupérer l'image**

```python
image = soup.find("img", alt="omg batman")

print(image["alt"])
print(urljoin(url, image["src"]))
display(Image(url=image["src"]))
```

**Récupérer le poème**

```python
poeme = soup.find("pre").get_text(strip=True)

print(poeme)
```
