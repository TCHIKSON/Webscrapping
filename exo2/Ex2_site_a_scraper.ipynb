import des éléments 

```python

import re
import requests
from bs4 import BeautifulSoup
```

url du site à scrper

```python
url = "https://www.pythonscraping.com/pages/warandpeace.html"
response = requests.get(url, timeout=10)
response.raise_for_status()
```

Soup

```python
soup = BeautifulSoup(response.text, "html.parser")
#print(soup.get_text())
```

**Récupérer ts les mots en vert**

```python

for element in soup.find_all("span", class_="green"):
    mots = re.findall(r"[A-Za-zÀ-ÿ]+(?:['’-][A-Za-zÀ-ÿ]+)?", element.get_text())
    
    for motVert in mots:
        print(motVert)
```

Récupérer ts les mots en rouge

```python

for element in soup.find_all("span", class_="red"):
    mots = re.findall(r"[A-Za-zÀ-ÿ]+(?:['’-][A-Za-zÀ-ÿ]+)?", element.get_text())
    
    for motRouge in mots:
        print(motRouge)
```

# Tous les mots rouges ou vert 

```python
elements_colores = soup.select("span.red, span.green")

mots_colores = []

for element in elements_colores:
    mots = re.findall(r"[A-Za-zÀ-ÿ]+(?:['’-][A-Za-zÀ-ÿ]+)?", element.get_text())
    mots_colores.extend(mots)
print(mots_colores)
```

**Afficher le premier mot :**

```python
print(mots_colores[0])
```

**Afficher les cinq premiers :**

```python
print(mots_colores[:5])
```

# Récupérer tous les mots qui correspondent à "the prince" en miniscule et coloré en vert ?

```python
the_prince = []

for element in soup.find_all("span", class_="green"):
    texte = element.get_text(" ", strip=True).lower()

    if texte == "the prince":
        the_prince.append(texte)

print(the_prince)
print(len(the_prince))
```
