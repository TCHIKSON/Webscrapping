```python
from IPython.display import Image, display
from urllib.parse import urljoin
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

url du site à scrper

```python
url = "https://www.nytimes.com/interactive/2017/06/23/opinion/trumps-lies.html"
response = requests.get(url, timeout=10)
response.raise_for_status()
print("Code HTTP :", response.status_code)
```

Soup

```python
soup = BeautifulSoup(response.text, "html.parser")
elements = soup.select("span.short-desc")
#print(soup.get_text())
print("Nombre d'éléments :", len(elements))
```

**Extraction des quatre informations demandées**

```python
donnees = []

for element in elements:
    date_element = element.find("strong")
    truth_span = element.find("span", class_="short-truth")

    lien = truth_span.find("a", href=True)

    date = date_element.get_text(strip=True).rstrip(",") + ", 2017"
    mensonge = element.contents[1].strip()
    explanation = truth_span.get_text(" ", strip=True).strip("() ")
    url = lien["href"]

    donnees.append([date, mensonge, explanation, url])

df = pd.DataFrame(
    donnees,
    columns=["date", "lie", "explanation", "url"]
)

df.head(10)
```
